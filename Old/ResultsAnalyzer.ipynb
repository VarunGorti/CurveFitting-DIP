```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import random

import torch
import torch.nn as nn
import torch.nn.functional as F

import math
import os
import time
import pandas as pd
```

```python
X = torch.load("/scratch/04703/sravula/UTAFSDataNew/NEW_DATA.pt")
F = torch.load("/scratch/04703/sravula/UTAFSDataNew/NEW_FREQUENCIES.pt")
```

```python
results_root = "/scratch/04703/sravula/dip_results/run_1"
```

```python
for i in range(7,8):
    x = X[i]
    f = F[i]
    
    path1 = os.path.join(results_root, "chip"+str(i), "1.pt")
    path2 = os.path.join(results_root, "chip"+str(i), "2.pt")
    path5 = os.path.join(results_root, "chip"+str(i), "5.pt")
    path10 = os.path.join(results_root, "chip"+str(i), "10.pt")
    
    x1 = torch.load(path1)
    x2 = torch.load(path2)
    x5 = torch.load(path5)
    x10 = torch.load(path10)
    
    if f is None or x.shape[-1] > 10:
        continue

    plt.figure(figsize=(12,9))
    for j in range(x.shape[-1]):
        plt.plot(f, x[0,0,:,j], label="Real " + str(j))
        plt.plot(f, x[0,1,:,j], label="Imag " + str(j))
    plt.xlabel("F (HZ)")
    plt.legend()
    plt.title("CHIP " + str(i) + " ACTUAL")
    plt.show()
    
    plt.figure(figsize=(12,9))
    for j in range(x.shape[-1]):
        plt.plot(f, x1[0,0,:,j], label="Real " + str(j))
        plt.plot(f, x1[0,1,:,j], label="Imag " + str(j))
    plt.xlabel("F (HZ)")
    plt.legend()
    plt.title("CHIP " + str(i) + " 1 PERCENT")
    plt.show()
    
    plt.figure(figsize=(12,9))
    for j in range(x.shape[-1]):
        plt.plot(f, x2[0,0,:,j], label="Real " + str(j))
        plt.plot(f, x2[0,1,:,j], label="Imag " + str(j))
    plt.xlabel("F (HZ)")
    plt.legend()
    plt.title("CHIP " + str(i) + " 2 PERCENT")
    plt.show()
    
    plt.figure(figsize=(12,9))
    for j in range(x.shape[-1]):
        plt.plot(f, x5[0,0,:,j], label="Real " + str(j))
        plt.plot(f, x5[0,1,:,j], label="Imag " + str(j))
    plt.xlabel("F (HZ)")
    plt.legend()
    plt.title("CHIP " + str(i) + " 5 PERCENT")
    plt.show()
    
    plt.figure(figsize=(12,9))
    for j in range(x.shape[-1]):
        plt.plot(f, x10[0,0,:,j], label="Real " + str(j))
        plt.plot(f, x10[0,1,:,j], label="Imag " + str(j))
    plt.xlabel("F (HZ)")
    plt.legend()
    plt.title("CHIP " + str(i) + " 10 PERCENT")
    plt.show()
```
