```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import random

import torch
import torch.nn as nn
import torch.nn.functional as F

import math
import time
import pandas as pd
```

```python
X = torch.load("/scratch/04703/sravula/UTAFSDataNew/NEW_DATA.pt")
F = torch.load("/scratch/04703/sravula/UTAFSDataNew/NEW_FREQUENCIES.pt")
```

```python
print(len(X))
print(len(F))
```

```python
for i, x in enumerate(X):
    if x is None:
        print(str(i) + ": NONE")
        print()
    else:
        num_unique = x.shape[-1]
        num_ports = int(0.5*(-1+np.sqrt(8*num_unique+1)))
        print(str(i) + " : " + str(list(x.shape)))
        print("NUM PORTS: " + str(num_ports))
```

```python
for i in range(len(X)):
    x = X[i]
    f = F[i]
    
    if f is None or x.shape[-1] > 10:
        continue

    plt.figure(figsize=(12,9))
    for j in range(x.shape[-1]):
        plt.plot(f, x[0,0,:,j], label="Real " + str(j))
        plt.plot(f, x[0,1,:,j], label="Imag " + str(j))
    plt.xlabel("F (HZ)")
    plt.legend()
    plt.title("CHIP " + str(i))
    plt.show()
```
